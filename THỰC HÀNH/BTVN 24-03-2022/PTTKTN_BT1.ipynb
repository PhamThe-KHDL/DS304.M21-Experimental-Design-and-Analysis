# Bài tập one-sample t-test


## Nhập dữ liệu

```r
d = c(8 , 18 , 11 , -3 , 11 , -2 , -2 , 10 , 8 , 13)
dbar = mean(d)
s = sd(d)
n = 10
df = 9
print(dbar)
print(s)
```

## Câu 1

### Kiểm định giả thuyết

- Đặt μ là mean của toàn bộ quần thể B-A
- H0: μ=0 (doanh số hai loại thật sự bằng nhau)
- H1: μ>0 (doanh số loại B thật sự lớn hơn loại A)

P(dbar>=7.2) = P(t>=3.1792)

```r
mu = 0
t = (dbar - mu)/(s/sqrt(n))
print(t)
```

```r
pt(t,df,lower.tail = FALSE)
```

Nếu H0 là đúng (mu=0) thì xác suất nhận được kết quả như hiện tại và extreme hơn P(dbar>=7.2) = 0.0056 < α (quá thấp để có thể tin có khả năng xảy ra) => bác bỏ H0 => chấp nhận H1 (doanh số B thật sự lớn hơn A)

### Trực quan hóa câu 1

```r
set_plot_dimensions <- function(width_choice, height_choice) {options(repr.plot.width=width_choice, repr.plot.height=height_choice)}
```

```r
Tvalues <- seq(-4,4,.1)
x <- Tvalues
y <- dt(Tvalues,9)

set_plot_dimensions(6, 4)
plot(x , y , type = "l" , xlab = "t" , ylab = "density(t)")
polygon(c(x[x>=t],min(x[x>=t])),c(y[x>=t],0),col="red")
```

## Câu 2

### Kiểm định giả thuyết

- Đặt μ là mean của toàn bộ quần thể B-A
- H0: μ=10 (chênh lệch doanh số B-A là 10 lon/tuần)
- H1: μ<10 (chênh lệch doanh số B-A ít hơn 10 lon/tuần)

```r
mu = 10
t = (dbar-mu)/(s/sqrt(n))
print(t)
```

P(dbar <= 7.2) = P(t <= -1.236)

```r
pt(t,df,lower.tail = TRUE)
```

Nếu H0 đúng (mu=10) thì xác suất thu được kết quả như hiện tại và extreme hơn P(dbar<=7.2) = 0.124 (không quá nhỏ để có thể coi như không thể xảy ra). Do đó ta không thể bác bỏ H0 (tức là xác nhận khả năng mu=10 là có thật). Do đó có thể cân nhắc đầu tư sản phẩm B.

### Trực quan hóa câu 2

```r
Tvalues = seq(-4,4,.1)
x = Tvalues
y = dt(Tvalues,9)

plot(x , y , type = "l" , xlab = "t" , ylab = "Density(t)")
polygon(c(x[x<=t],max(x[x<=t])), c(y[x<=t],0), col="red")
```

## Câu 3

### Tính khoảng tin cậy

```r
t025 = abs(qt(0.025,9))
# tinh khoang tin cay [lbound,ubound]
lbound = dbar-t025*s/sqrt(n)
ubound = dbar+t025*s/sqrt(n)
sprintf("CI: [%f ; %f]", lbound, ubound)
```

### Trực quan hóa câu 3

```r
m = dbar+T.values*s/sqrt(n)
plot(m , y , type = "l" , xlab = "Actual mean" , ylab = "Probability density")
polygon(c(m[m<=lbound], max(m[m<=lbound])), c(y[m<=lbound], 0), col="red")
polygon(c(m[m>=ubound], min(m[m>=ubound])), c(y[m>=ubound], 0), col="red")
```
