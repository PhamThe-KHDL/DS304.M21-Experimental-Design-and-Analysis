# Bi tp one-sample t-test


## Nhp d liu

```r
d = c(8 , 18 , 11 , -3 , 11 , -2 , -2 , 10 , 8 , 13)
dbar = mean(d)
s = sd(d)
n = 10
df = 9
print(dbar)
print(s)
```

## Cu 1

### Kim nh gi thuyt

- t  l mean ca ton b qun th B-A
- H0: =0 (doanh s hai loi tht s bng nhau)
- H1: >0 (doanh s loi B tht s ln hn loi A)

P(dbar>=7.2) = P(t>=3.1792)

```r
mu = 0
t = (dbar - mu)/(s/sqrt(n))
print(t)
```

```r
pt(t,df,lower.tail = FALSE)
```

Nu H0 l ng (mu=0) th xc sut nhn c kt qu nh hin ti v extreme hn P(dbar>=7.2) = 0.0056 <  (qu thp  c th tin c kh nng xy ra) => bc b H0 => chp nhn H1 (doanh s B tht s ln hn A)

### Trc quan ha cu 1

```r
set_plot_dimensions <- function(width_choice, height_choice) {options(repr.plot.width=width_choice, repr.plot.height=height_choice)}
```

```r
Tvalues <- seq(-4,4,.1)
x <- Tvalues
y <- dt(Tvalues,9)

set_plot_dimensions(6, 4)
plot(x , y , type = "l" , xlab = "t" , ylab = "density(t)")
polygon(c(x[x>=t],min(x[x>=t])),c(y[x>=t],0),col="red")
```

## Cu 2

### Kim nh gi thuyt

- t  l mean ca ton b qun th B-A
- H0: =10 (chnh lch doanh s B-A l 10 lon/tun)
- H1: <10 (chnh lch doanh s B-A t hn 10 lon/tun)

```r
mu = 10
t = (dbar-mu)/(s/sqrt(n))
print(t)
```

P(dbar <= 7.2) = P(t <= -1.236)

```r
pt(t,df,lower.tail = TRUE)
```

Nu H0 ng (mu=10) th xc sut thu c kt qu nh hin ti v extreme hn P(dbar<=7.2) = 0.124 (khng qu nh  c th coi nh khng th xy ra). Do  ta khng th bc b H0 (tc l xc nhn kh nng mu=10 l c tht). Do  c th cn nhc u t sn phm B.

### Trc quan ha cu 2
